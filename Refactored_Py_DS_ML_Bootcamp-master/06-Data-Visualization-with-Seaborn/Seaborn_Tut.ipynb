#Distribution plot
#for numerical values in a dataframe
#distribution plot(dataframe colmn), jointplot(x,y,data), rugplot(dataframe colmn), pairplot(dataframe)

```python
sns.distplot(tips['tip'],kde=False)
```

```python
sns.rugplot(tips['tip'])
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')#kind options (hex,reg,kde)
```

```python
sns.pairplot(tips,hue='smoker')
```

#categorical plot
#mostly for showing rship between non-numerical colmuns and a numerical column
barplot, countplot, boxplot,violinplot,

```python
sns.barplot(x='sex',y='total_bill',data = tips)
```

```python
sns.countplot(x='time',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips, jitter=True,hue='sex')
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips)
```
