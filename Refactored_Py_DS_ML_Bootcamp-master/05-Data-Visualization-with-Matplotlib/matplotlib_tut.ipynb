```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
plt.plot(x,y)
plt.xlabel('X LBL')
plt.ylabel('Y LBL')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X label')
axes.set_ylabel('y label')
axes.set_title('Graph')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.6,0.75,0.2,0.1])
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
#for i in axes:
#    i.plot(x,y)
axes[0].plot(x,y)
axes[0].set_title('first')
axes[1].plot(y,x)

    
plt.tight_layout()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label = 'X By 2',color='green',marker='*',markerfacecolor='red',markeredgecolor='red')
ax.plot(x,x**3,label = 'X By 3',marker='*',markerfacecolor='red',markeredgecolor='red')

ax.legend()
plt.show()
```
